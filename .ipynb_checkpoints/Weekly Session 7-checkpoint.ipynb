We also need to be able to generate matrices. We have done this two ways:

But, traditionally, you would start from an array which already has been created, with the right dimensions, and then you would fill in the entries one by one. 

<div class=answer>

So, in the spirit of ```numpy.zeros```, write a function zeros which takes a tuple ```(m,n)``` as its argument and returns an $m\times n$ matrix of zeros.

````python
#your code here
def zeroes (m,n):
    a = [0] * n
    for i in range (n):
        a[i]= [0] * m
    return a

m = 5
n = 2
print (zeroes (m,n))
````


While we are at it, let's write a function ```identity()```, which makes an $n\times n$ identity matrix.  If you need a reminder/lesson, the identity matrix is: 
$$
I_3 = \left[ \begin{array}{cccc}
1 & 0 & 0  \\
0 & 1 & 0 \\
0 & 0 & 1 \\ \end{array} \right]
$$

Since the identity matrix is square, you need just one number n as an argument.  Also, be sure NOT to duplicate work you've already done.  Be lazy.
<div class=answer>

````python
#your code here
def identity (n):
    a = zeroes(n,n)
    for i in range (n):
        a[i][i] = 1
    return a

n = 5
print(identity(n))
````

### Matrix Arithmetic

As you either already know from other physics/math courses, or will shortly, one of the basic things you can do to a ```A``` is multiply it by a scalar ```k```. That is, the $(i,j)$th entry of the product matrix, let's call it ```C```, is defined by
$$
C_{i,j} = k\ A_{i,j}
$$
<div class=answer>
Write a function ```scalar_mult``` which multiplies a matrix by a scalar, returning a new matrix.
<div/>

````python
#your code here
def scalar_mult (matrix,scalar):
    rows, columns = dimension(matrix)
    for i in range (rows):
        for j in range (columns):
            matrix[i][j] *= scalar
    return matrix

matrix = [[1,2],[3,4]]
scalar = 5
print(scalar_mult(matrix,scalar))
````

If we have two matrices ```A``` and ```B``` of the same dimensions, we also can compute their sum.
That is, in terms of the entries, the $(i,j)$ th entry of the sum ```C``` is defined by
$$
C_{ij} =  A_{ij} + B_{ij} 
$$

<div class=answer>
Write a function ```add()```, which takes two matrices, and returns their sum. It should check that the dimensions of the two matrices A and B are compatible, and generate an error message (that is, an exception) if not.
<div/>

````python
#your code here

def add (matrixA, matrixB):
    rowA, columnA = dimension(matrixA)
    rowB, columnB = dimension(matrixB)
    if (rowA != rowB or columnA != columnB):
        raise ValueError ("Matrix Dimensions are not compatible")
        return
    C = zeroes (columnA, rowA)
    for i in range (rowA):
        for j in range (columnA):
            C[i][j] = matrixA[i][j] + matrixB [i][j]
    return C

print (add (A,B))
````

You knew this was coming, didn't you. The product of an $m \times r$ matrix $A$ and an $r \times n$ matrix $B$ is an $m \times n$ matrix with entries given by
$$
C_{ik} = \sum_{j=1}^r A_{ij} B_{jk}
$$

Write a function ```mult``` which multiplies two matrices, returning a new matrix. It should check that the dimensions of the two matrices $A$ and $B$ are compatible, and generate an exception if not. Try it out with the example product shown above.
<div/>

````python
#your code here

def mult (matrixA, matrixB):
    rowA, columnA = dimension(matrixA)
    rowB, columnB = dimension(matrixB)
    if (columnA != rowB):
        raise ValueError ("Matrix Dimensions are not compatible")
        return
    C = zeroes (columnB, rowA)
    for i in range (rowA):
        for j in range(columnB):
            for k in range(columnA):
                C[i][j]+= A[i][k] * B[k][j]
    return C

A = [[1,2,-1],[3,0,2]]
B = [[1,2],[7,0],[-1,3]]

print("C=",mult(A,B))
````

## Exercise 3
Do your interactive practice for Wednesday's class here.  Specifically code up Newman's Example 6.1, but encapsulate the Gaussian elimination algorithm into a function you may call (in Exercise 4).

````python
#your code here

def Gaussian_Elimination (A, v):
    """Performs a Gaussian Elimination of a Matrix and a resultant vector to produce the multiplicative vector"""
    N = len(v)
    for m in range(N):
        div = A[m,m]
        A[m,:] /= div
        v[m] /= div
        
        for i in range(m+1,N):
            mult = A[i,m]
            A[i,:] -= mult * A[m,:]
            v[i] -= mult * v[m]
    
    x = np.empty(N)
    for j in range (N-1,-1, -1):
        x[j] = v[j]
        for k in range (j+1,N):
            x[j] -= A[j,k] * x[k]
    return x

A = np.array([[3, 2, 4, 1],
             [3, 4, -1, -1],
             [1, -4, 1, 5],
             [2, -2, 1, 3]],float)
v = np.array([-4, 3, 9, 7],float)
print(Gaussian_Elimination(A,v))
````

````python
for i in range (8):
    # Constants
    N = 10
    C = 1.0
    m = 1.0
    k = 6.0
    omega = i+1
    alpha = 2 * k - m * omega * omega
    title = "System with ", omega, "Frequency"

    # Initialize the Arrays
    A = np.zeros([N,N],float)
    for i in range (N-1):
        A[i,i] = alpha
        A[i, i+1] = -k
        A[i+1, i] = -k
    
    A[0,0] = alpha - k
    A[N-1, N-1] = alpha - k
    v = np.zeros(N, float)
    v[0] = C

    # Gaussian Elimination
    x = Gaussian_Elimination(A, v)
    plt.plot(x,)
    plt.plot(x, "ko")
    plt.title(title)
    plt.show()
````

Increasing the frequency here makes the graph begin to be less smooth, and leads to a dampening effect. Interestingly, this pattern actually continues up through a frequency of 24, which seems counter to the idea of fundamental frequencies. It may simply be that the number of masses makes it too hard to graph. Due to the way the equation was defined, having omegas over a certain value resulted in alpha being negative. This leads to the damping factor seen above a certain threshold. By increasing k, we can increase omega as well. This relation is given by:
$$ 2k > m\omega^{2} $$

This relationship is demonstrated below.

````python
# Constants
N = 20
C = 1.0
m = 1.0
k = 14.0
omega = 5.0
alpha = 2 * k - m * omega * omega

# Initialize the Arrays
A = np.zeros([N,N],float)
for i in range (N-1):
    A[i,i] = alpha
    A[i, i+1] = -k
    A[i+1, i] = -k
    
A[0,0] = alpha - k
A[N-1, N-1] = alpha - k
v = np.zeros(N, float)
v[0] = C

# Gaussian Elimination
x = Gaussian_Elimination(A, v)
plt.plot(x)
plt.title("System w/ k = 14 & omega = 5")
plt.show()
````

As can be seen above, a system of 20 test masses behaved in a much more predicted fashion when the k was increased to compensate for the omega. Compared to the default case of omega=5, this behaved much smoother (omega=5 was the most apparent step into dampening, even though the function goes negative at omega=4).

````python
for i in range (5,50,5):
    # Constants
    N = 10
    C = 1.0
    m = 1.0
    k = i
    omega = 2.0
    alpha = 2 * k - m * omega * omega

    # Initialize the Arrays
    A = np.zeros([N,N],float)
    for i in range (N-1):
        A[i,i] = alpha
        A[i, i+1] = -k
        A[i+1, i] = -k
    
    A[0,0] = alpha - k
    A[N-1, N-1] = alpha - k
    v = np.zeros(N, float)
    v[0] = C

    # Gaussian Elimination
    x = Gaussian_Elimination(A, v)
    plt.plot(x,)
    plt.plot(x, "ko")
    plt.show()
````

Altering the spring constant caused the system to move into something closer to a single half-wavelength. With each increase in k, the function gets closer and closer to being almost completely linear. Notably, these were all for the same omega and number of masses.

````python
for i in range (3,10,1):
    # Constants
    N = 10
    C = 1.0
    m = i+1
    k = 6.0
    omega = 2.0
    alpha = 2 * k - m * omega * omega
    title = "System with masses of", i

    # Initialize the Arrays
    A = np.zeros([N,N],float)
    for i in range (N-1):
        A[i,i] = alpha
        A[i, i+1] = -k
        A[i+1, i] = -k
    
    A[0,0] = alpha - k
    A[N-1, N-1] = alpha - k
    v = np.zeros(N, float)
    v[0] = C

    # Gaussian Elimination
    x = Gaussian_Elimination(A, v)
    plt.plot(x,)
    plt.plot(x, "ko")
    plt.title(title)
    plt.show()
````

Varying m produced a similar effect to varying omega, but at a much slower rate, which makes sense. Omega is raised to a second power, so it takes less to make alpha negative compared to m, which is only to a first power. It does still show the same damping effect as mass increases. Physically, this makes sense, as mass is often associated with the ability to resist motion by a force, and thus the larger the masses, the less the spring force will move them.
